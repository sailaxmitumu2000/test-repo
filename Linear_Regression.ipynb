Load The needed libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

Load the data set

```python
data_path="/Salary_dataset.csv"
```

```python
df=pd.read_csv(data_path)
```

```python
df.head()
```

```python
# Define input and output
X=df[['YearsExperience']]
Y=df[['Salary']]
```

```python
# Create and train the model
model=LinearRegression()
model.fit(X,Y)
```

```python
# Print the model equation
y_pred=model.predict(X)
```

```python
print(f"\nModel Intercept (b): {model.intercept_[0]:.2f}")
print(f"Model Coefficient (m): {model.coef_[0][0]:.2f}")
print(f"Equation: Salary = {model.coef_[0][0]:.2f} * YearsExperience + {model.intercept_[0]:.2f}")
```

```python
years_of_experience = np.array([[2]])
predicted_salary = model.predict(years_of_experience)
```

```python
# Visualize
plt.figure(figsize=(8, 5))
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression Line')
plt.title("Simple Linear Regression - Salary vs Years of Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.grid(True)
plt.show()
```

```python
mse = mean_squared_error(Y, y_pred)
rmse = mse ** 0.5
r2 = r2_score(Y, y_pred)

print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (R²): {r2:.2f}")
```

# Things to know about Linear Regression

**Key Assumptions**
For linear regression to work well, these assumptions should hold:

Linearity: Relationship between inputs and output is linear.

Independence: Data points are independent.

Homoscedasticity: The variance of residuals (errors) is constant for all input values.

Normality: Residuals are normally distributed (important for statistical tests).

No multicollinearity: For multiple features, the features shouldn’t be highly correlated.



**How to Evaluate the Model?**
R-squared (R²): Percentage of variance explained by the model (0 to 1). Higher means better fit.

Adjusted R-squared: Adjusts for number of features; useful for multiple regression.

Mean Squared Error (MSE): Average squared difference between predicted and actual values. Lower is better.

Root Mean Squared Error (RMSE): Square root of MSE — in same units as output.

Mean Absolute Error (MAE): Average absolute difference — less sensitive to outliers.
